```python
import pandas as pd 
df = pd.read_json("../data/cocktail_dataset.json")
df
```

```python
df['alcoholic'].value_counts()
```

```python
def update_tags(row):
  tags = row['tags'] if isinstance(row['tags'], list) else []
  if "Alcoholic" not in tags:
    tags.append("Alcoholic")
  return tags

df['tags'] = df.apply(update_tags,axis=1)

df
```

```python
df['ingredients'][0]
```

```python
def extract_ingredients_with_measures(arr):
  ingredients = []
  for ingredient in arr:
    ingredient_name = ingredient.get('name')
    measure = ingredient.get('measure', '')  
    if measure:
      ingredients.append(f"{ingredient_name} {measure.strip()}")
    else:
      ingredients.append(ingredient_name)
  return ingredients

df['ingredients'] = df['ingredients'].apply(lambda x : extract_ingredients_with_measures(x))

```

```python
def build_text(row):
    text = row['name'] + '\n\n' + 'Category: ' + row['category'] + '\n\n' + 'Glass: ' + row['glass'] + '\n\n' + 'Tags: ' + ', '.join(row['tags']) + '\n\n' + 'Instructions: ' + row['instructions'] + '\n\n' + 'Ingredients: ' + ', '.join(row['ingredients'])
    return text

df['text'] = df.apply(build_text,axis=1)
```
